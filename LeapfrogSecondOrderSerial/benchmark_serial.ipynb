```julia
#for reading in ascii files
include("../rw_functions.jl");
#contains the leapfrog serial rountine
include("leapfrog_second_order_serial.jl")
```

```julia
setup=read_setup_from_file("test20particle");
mass=get_mass(setup);
initialState=get_initialState(setup);
```

```julia
nsteps=100000;
stepsize=0.5;
grav_constant=0.1;

tic();
pos=integrate_leapfrog_ser(initialState,mass,nsteps,stepsize,grav_constant);
toc();
```

```julia
using PyPlot;
numparticles = size(initialState)[1]; #the number of rows of state
linesize=mass./(maximum(mass))
for i=1:numparticles
    x=transpose(pos[i,:,1]);
    y=transpose(pos[i,:,2]);
    #z=transpose(pos[i,:,3]);
   #print(x)
    plot(x,y,linewidth=10*linesize[i])
end
#xlim(-100000,100000)
#ylim(-100000,100000)
```

```julia
function comp_mass_center(d::Array{Float64,1},mass::Array{Float64,1})
    #component center of mass calculation
    return sum(d.*mass)/sum(mass)
end
```
